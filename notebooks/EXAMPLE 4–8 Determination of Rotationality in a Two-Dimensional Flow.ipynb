```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def V_uv(x, y):
    u = x**2
    v = -2*x*y -1
    return u, v
```

```python
dt = 0.25

scale = 0.20

pos_0 = np.array([
    [1, 4+scale],
    [1+scale, 4+scale],
    [1+scale, 4],
    [1+scale, 4-scale],
    [1, 4-scale],
    [1-scale, 4-scale],
    [1-scale, 4],
    [1-scale, 4+scale],
    
    [1, 4+scale],
])

ds_1 = np.array([np.array([*V_uv(*pos)])*dt for pos in pos_0])
pos_1 = pos_0 + ds_1

ds_2 = np.array([np.array([*V_uv(*pos)])*dt for pos in pos_1])
pos_2 = pos_1 + ds_2

plt.plot(pos_0[:,0], pos_0[:,1], "-o")
plt.plot(pos_1[:,0], pos_1[:,1], "-o")
plt.plot(pos_2[:,0], pos_2[:,1], "-o")
plt.axis('equal')
plt.xlim([0, 5])
plt.ylim([0, 5])
plt.grid()
```
